# Gender Distribution

```python
plt.figure(figsize=(5,5))
ax=sns.countplot(data=df,x="Gender") #This will plot the ColuntPlot
ax.bar_label(ax.containers[0])  # This will show the count of male and female
plt.title("Gender Distribution")
plt.show()
```

# From The above chart we have analysed that :
# The number of females in the data is more than the number of females

```python
gb=df.groupby("ParentEduc").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"}) #This will Group by parent education related to students score
print(gb)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb,annot= True) #annot will show the values inside the shades
plt.title("Relationship between parent Education and Student Score")
plt.show() #darker shade will show lower value and lighter shade will show higher value
```

# from the Above chart we have concluded that the education of parent have a good impact on their score

```python
gb1 =df.groupby("ParentMaritalStatus").agg({"MathScore":"mean","ReadingScore":"mean","WritingScore":"mean"}) #This will Group by parent education related to students score
print(gb1)
```

```python
plt.figure(figsize=(5,5))
sns.heatmap(gb1,annot= True) #annot will show the values inside the shades
plt.title("Relationship between parent Martial Status and Student Score")
plt.show()
```

# From the above chart,We have analyzed that the parent martial status have negligible impact on student scores

# Distribution Of Ethnic Group

```python
GroupA=df.loc[(df["EthnicGroup"]=="group A")].count()
GroupB=df.loc[(df["EthnicGroup"]=="group B")].count()
GroupC=df.loc[(df["EthnicGroup"]=="group C")].count()
GroupD=df.loc[(df["EthnicGroup"]=="group D")].count()
GroupE=df.loc[(df["EthnicGroup"]=="group E")].count()

l=["Group A","Group B","Group C","Group D","Group E"] #this will create labels for pie chart
mlist=[GroupA["EthnicGroup"],GroupB["EthnicGroup"],GroupC["EthnicGroup"],GroupD["EthnicGroup"],GroupE["EthnicGroup"]] #we have created list to pass it will easier the task
plt.pie( mlist,labels=l,autopct='%1.1f%%')
plt.title("Distribition of Ethnic Groups")
plt.legend()
plt.show() 
```
